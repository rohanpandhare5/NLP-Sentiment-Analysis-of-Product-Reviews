```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('HousingData.csv').drop(columns=['date','street','city','statezip','country'])
print(df.head())
```

```python
print(df.tail()) #shows last/tail values of dataset
```

```python
print(df.describe()) #To describe propertires like min max mean count for the data
```

```python
print(df.isnull().sum())   #checks the null values. is null tells True/False .sum gives total null values
```

```python
df.columns #checks the columns
```

```python
sns.heatmap(df.corr(), annot=True) #Gives correlation matrix between the columns
```

```python
x = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated']]

y = df[['price']]

print(x,y)
```

```python
train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05, random_state=66) #random state gives a particular set of random data whereas if we didnt put anything then it will show new random numbers everytime
```

```python
x_train
```

```python
lregr = LinearRegression()
lregr.fit(x_train, y_train)
```

```python
lregr.coef_[0] #here we are only seeing coefficients ie only b values in y=b1x1+b2x2+b3x3 .....+bnxn
```

```python
coef_df = pd.DataFrame(lregr.coef_[0],x_train.columns)  #here we are creating the data frame also including the column names
coef_df
```

```python
y_predict = lregr.predict(x_test)
y_predict
```

```python
plt.scatter(y_test, y_predict,edgecolor="black")
plt.xlim(0,3000000)
plt.ylim(0,3000000)
```
