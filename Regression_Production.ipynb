```python
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = genfromtxt('production_new.csv', delimiter=',')
df
```

```python
x= array(df[:,0]).reshape(-1,1)
y=array(df[:,1]).reshape(-1,1)
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
plt.scatter(x,y)
```

```python
regr= LinearRegression()
regr.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train)
```

```python
regr.predict([[1300]])
```

```python
y_predict_test = regr.predict(x_test)
y_predict_train = regr.predict(x_train)

plt.plot(x_train,y_predict_train, color="Green")
plt.plot(x_test, y_predict_test)

plt.scatter(x_test,y_test)
```
